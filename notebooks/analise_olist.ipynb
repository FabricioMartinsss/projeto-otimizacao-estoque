# Otimização de Estoque com Análise de Dados

Este projeto tem como objetivo analisar dados reais de e-commerce para identificar padrões de vendas, sazonalidade e comportamento de compra dos clientes.

O foco principal da análise é entender como os dados podem ser utilizados para **otimizar o controle de estoque**, reduzir produtos parados e melhorar a eficiência das vendas.

As análises foram desenvolvidas utilizando **SQL, Python e Pandas**, com os dados armazenados em um banco SQLite construído a partir do dataset Brazilian E-Commerce (Olist).

Neste notebook estão presentes as etapas de exploração dos dados, consultas em SQL e extração de insights voltados para tomada de decisão em um cenário real de e-commerce.
